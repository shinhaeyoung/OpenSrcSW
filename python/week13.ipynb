```python
# 데이터 시각화 실습1: 생일 기온 시각화

# 자신 생일의 서울 지역 최고 기온과 최저 기온 시각화 (10년치 데이터)

import csv
import matplotlib.pyplot as plt

f = open('weather_10years.csv')
data = csv.reader(f)
next(data)
high = []
low = []
for row in data:
    if row[-1] != '' and row[-2] != '':
        if row[0].split('-')[1] == '02' and row[0].split('-')[2] == '08':
            high.append(float(row[-1]))
            low.append(float(row[-2]))
f.close()

plt.title('my birthday weather')
plt.plot(high, 'firebrick', label='Hightest')
plt.plot(low, 'lightskyblue', label='Lowest')
plt.legend()
plt.show()
```

```python
# 데이터 시각화 실습2: 최고 기온 시각화

# 서울지역의 1월과 8월 최고 기온 데이터를 히스토그램으로 시각화 (10년치 데이터)

import csv
import matplotlib.pyplot as plt

f = open('weather_10years.csv')
data = csv.reader(f)
next(data)
jan = []
aug = []
for row in data:
    month = row[0].split('-')[1]
    if row[-1] != '':
        if month == '01':
            jan.append(float(row[-1]))
        elif month == '08':
            aug.append(float(row[-1]))

plt.title('Jan Aug Highest Temperature')
plt.hist(jan, bins=100, color='lightskyblue', label='January')
plt.hist(aug, bins=100, color='firebrick', label='August')
plt.legend()
plt.show()
```
