```python
# 공공데이터 실습1: 서울의 일교차

# 기온 공공 데이터를 이용하여 서울에서 일교차가 가장 큰 날, 
# 최저기온이 가장 낮은 날, 최고기온이 가장 높은 날 출력

# header => 날짜 지점 평균기온(℃) 최저기온(℃) 최고기온(℃)

import csv
f = open('weather.csv')
data = csv.reader(f)
header = next(data)

maxTemp = -999
minTemp = 999
diffTemp = 0 
maxDate = minDate = diffDate = ''

for row in data:
    if row[-1] == '': # 최고기온 누락시
        row[-1] = -999
    if row[-2] == '': # 최저기온 누락시
        row[-2] = 999
        
    row[-1] = float(row[-1]) # 최고기온
    row[-2] = float(row[-2]) # 최저기온
    if maxTemp < row[-1]:
        maxDate = row[0]
        maxTemp = row[-1]
    if minTemp > row[-2]:
        minDate = row[0]
        minTemp = row[-2]
        
    diff = row[-1] - row[-2] # 일교차
    if diffTemp < diff:
        diffDate = row[0]
        diffTemp = diff
f.close()
print("""서울의 일교차가 가장 큰 날: {0:s}
서울의 최저 기온이 가장 낮은 날: {1:s}
서울의 최고 기온이 가장 높은 날: {2:s}""".format(diffDate, minDate, maxDate))
```

```python
# 공공데이터 실습2: 지하철 이용객

# '지하철 노선별 역별 이용현황'데이터를 이용하여
# 노선별 가장 이용객이 많은 3개의 역과 가장 적은 3개의 역 출력

# header => 사용월 호선명 역ID 지하철역 승차승객수 하차승객수

import pandas as pd
data = pd.read_csv('subwayuser.csv', encoding='cp949')
label = ['호선명', '지하철역', '승차승객수', '하차승객수', '전체승객수']

# 승차승객수
data[label[2]] = data[label[2]].str.replace(pat=',', repl='').astype(int)
# 하차승객수
data[label[3]] = data[label[3]].str.replace(pat=',', repl='').astype(int)
# 전체승객수
data[label[-1]] = data[label[2]] + data[label[3]]
data = data[[label[0], label[1], label[-1]]]

# 호선별 전체승객수 상위 3개
def top(data):
    return data.sort_values(by=label[-1], ascending=False)[:3]
top3 = data.groupby(label[0]).apply(top)
topList = top3.values.tolist()
#print(topList)

# 호선별 전체승객수 하위 3개
def bottom(data):
    return data.sort_values(by=label[-1], ascending=True)[:3]
bottom3 = data.groupby(label[0]).apply(bottom)
bottomList = bottom3.values.tolist()
#print(bottomList)

for i in range(0, len(topList), 3):
    print("<{0:s}>\n이용객이 가장 많은 역 순위\n1. {1:s}({2:d})\n2. {3:s}({4:d})\n3. {5:s}({6:d})"
          .format(topList[i][0], topList[i][1], topList[i][2], 
                  topList[i+1][1], topList[i+1][2], topList[i+2][1], topList[i+2][2]))
    print("이용객이 가장 적은 역 순위\n1. {0:s}({1:d})\n2. {2:s}({3:d})\n3. {4:s}({5:d})\n"
          .format(bottomList[i][1], bottomList[i][2], 
                  bottomList[i+1][1], bottomList[i+1][2], bottomList[i+2][1], bottomList[i+2][2]))

```
